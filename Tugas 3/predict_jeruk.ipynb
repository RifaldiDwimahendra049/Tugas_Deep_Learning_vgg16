```python
Trained_model = load_model('vgg16_jeruk.h5')
```

```python
img = Image.open("Jeruk hijau.jpg")

img = img.resize((224, 224))
img = np.asarray(img)
plt.imshow(img)
img = np.expand_dims(img, axis=0)


output = Trained_model.predict(img)

if output[0][0] > output[0][1]:
    print("Jeruk matang")
    print("akurasi :",round(output[0][0]*100, 2),"%")
else:
    print('Jeruk belum matang ')
    print("akurasi :",round(output[0][1]*100,2),"%")
```
